```python
import cv2
```

```python
img = cv2.imread("scenery.png")
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
#inverted_img = 255- img
#cv2_imshow(inverted_img)
```

```python
inverted_gray_img = 255- gray_img
cv2_imshow(inverted_gray_img)
```

```python
blurred_img = cv2.GaussianBlur(inverted_gray_img,(21,21),0)
cv2_imshow(blurred_img)
```

```python
inverted_blur_img = 255 - blurred_img
cv2_imshow(inverted_blur_img)
```

```python
sketch_img = cv2.divide(gray_img, inverted_blur_img, scale=256.0)
#cv2_imshow(sketch_img)

darkened_sketch = cv2.convertScaleAbs(sketch_img, alpha=0.8, beta=0)  # Reduce contrast (brightness_factor as alpha)
cv2_imshow(darkened_sketch)
```

## 1. LOAD THE PICTURE USING CV2
## 2. CONVERT THE COLORED PIC INTO GRAYSCALE
## 3. INVERT THE GRAYSCALE IMAGE
## 4. BLUR THE INVERTED GRAYSCALE IMAGE
## 5. AGAIN INVERT THE BLURRED IMAGE
## 6. AT LAST , USING DIVIDE FUNCTION OF CV2 , CREATE A SKETCH OF THE PIC.
